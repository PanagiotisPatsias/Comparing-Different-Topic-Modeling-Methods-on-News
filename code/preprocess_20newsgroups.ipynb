```python
from sklearn.datasets import fetch_20newsgroups
import pandas as pd

# Load the dataset
newsgroups_data = fetch_20newsgroups(subset='all')
texts = newsgroups_data.data  # List of texts
labels = newsgroups_data.target  # List of target labels
label_names = newsgroups_data.target_names  # Names of the categories
# Convert to a DataFrame for easier handling
df = pd.DataFrame({'text': texts, 'label': labels})
df['label_name'] = df['label'].apply(lambda x: label_names[x])

df[['label_name']].value_counts()
```

```python
# Example manual grouping of the original 20 categories into 6 groups
label_groups = {
    'comp.graphics': 'Technology',
    'comp.os.ms-windows.misc': 'Technology',
    'comp.sys.ibm.pc.hardware': 'Technology',
    'comp.sys.mac.hardware': 'Technology',
    'comp.windows.x': 'Technology',

    'rec.autos': 'Vehicles',
    'rec.motorcycles': 'Vehicles',

    'rec.sport.baseball': 'Sport',
    'rec.sport.hockey': 'Sport',

    'sci.crypt': 'Science',
    'sci.electronics': 'Science',
    'sci.med': 'Science',
    'sci.space': 'Science',

    'talk.politics.guns': 'Politics',
    'talk.politics.mideast': 'Politics',
    'talk.politics.misc': 'Politics',
    'talk.religion.misc': 'Religion',

    'alt.atheism': 'Religion',
    'misc.forsale': 'Miscellaneous',
}

# Create a new column in the DataFrame for the broader group
df['group'] = df['label_name'].map(label_groups)

# Display the new DataFrame with broader groups
df[['group']].value_counts()
```
